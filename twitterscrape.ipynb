```python
#!/usr/bin/env python3
# -*- coding: utf-8 -*-
"""
Created on Sun Dec 12 04:30:00 2021

@author: travis


"""


#This will probably require beautiful soup
import argparse
import csv
from bs4 import BeautifulSoup
#import json
import os
from re import search
import requests
import sys
import time

def scrape_stats (player):    
    counter = 0
    start = time.perf_counter()

    with open(player, 'w', encoding = 'utf-8', newline = '\n') as csvfile:      
        #years = soup.findAll('a', text = re.compile(\d{4}-\d{2}))
        years = ['https://www.hockey-reference.com/players/j/jonesma03/gamelog/2019', 'https://www.hockey-reference.com/players/b/bitetan01/gamelog/2019']
        for year in years:
            year = requests.get(year)
            year = BeautifulSoup(year.text, 'lxml')
            #print('hi')
            table = year.find('tbody')
            rows = table.findChildren(['tr'])
            headers = ['date', 'goals', 'assists', '+/-', 'pim', 'shots', 'shot%', 'TOI', 'hits', 'blocks', 'faceoff %']
            headers = '\t'.join(headers)
            print(headers)
            for row in rows:
                cols = row.find_all('td')
                cols = [x.text.strip() for x in cols]
                out = []
                for count, col in enumerate(cols):
                    if count == 0:
                        out.append(col)
                    elif count == 7:
                        out.append(col)
                    elif count == 8:
                        out.append(col)
                    elif count == 10:
                        out.append(col)
                    elif count == 11:
                        out.append(col)
                    elif count == 19:
                        out.append(col)
                    elif count == 20:
                        out.append(col)
                    elif count == 22:
                        out.append(col)
                    elif count == 23:
                        out.append(col)
                    elif count == 24:
                        out.append(col)
                    elif count == 27:
                        out.append(col)
                
            

                out = '\t'.join(out)
                print(f'{out} \n')
        print('year complete!')
        
        end = time.perf_counter()
        runtime = end-start
        #print(f'{len(rows)} games scraped in {runtime} seconds')
        
def find_year(profile):
    counter = 0
    gmlogs = []
    profile = 'https://www.hockey-reference.com/players/j/jonesma03.html'
    player = requests.get(profile)
    soup = BeautifulSoup(player.text, 'lxml')
    for a in soup.find_all('a', href=True):
        counter += 1
        addy = a['href']
        addy = f'https://www.hockey-reference.com{addy}'
        if search('gamelog/\d{4}', addy):
            if addy in gmlogs:
                continue
            else:
                gmlogs.append(addy)
    print(gmlogs)
    print(f'{counter} urls found')
    
'''
parser = argparse.ArgumentParser(description = 'Player Handle')
parser.add_argument('--player', dest='player', type=str, help="NHL player")
arg = parser.parse_args()
'''
prof = 'https://www.hockey-reference.com/players/j/jonesma03.html'
player = requests.get('https://www.hockey-reference.com/players/j/jonesma03.html')
soup = BeautifulSoup(player.text, 'lxml')

#player = str(sys.argv[2])

player = input('player name:')

directory = 'data/'

#parent_dir = '../'
parent_dir = './'

path = os.path.join(parent_dir, directory)


#query = f'from:{handle} since:2005-01-01'
out_file = path + player + '.csv'
print(out_file)


if os.path.exists(path):
    print(f'Scraping statistical history for {player}')
    #scrape_stats(out_file)
    find_year(prof)
else:
    print('Creating data directory')
    print(f'Scraping statistical history for {player}')
    os.mkdir(path)
    scrape_stats(out_file)
```
