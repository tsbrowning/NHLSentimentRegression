```python
scraper = snscrape.modules.twitter.TwitterSearchScraper('from:ovi8 since:2005-01-01')
with open('example.txt', 'w', encoding = 'utf-8') as e:
    for tweet in scraper.get_items():
        tweet = str(tweet.json())
        json_object = json.loads(tweet)
        line = []
        c = str(json_object['content'])
        d = str(json_object['date'])
        line.append(c)
        line.append(d)
        l = ' * '.join(line)
        e.write(l)
        e.write('\n')
        print(l)
```
